### 01_Data_Loading_and_Merging.ipynb

```python
import pandas as pd
import pyarrow.parquet as pq
import os
import glob
```

## Load & Merge the files

```python
raw_data_path = "data/raw/*.parquet"
```

```python
parquet_files = glob.glob(raw_data_path)
print(f"Found {len(parquet_files)} files to merge.")
```

```python
dfs = []
for file in parquet_files:
    print(f"Processing {file} ...")
    table = pq.read_table(file)
    df_temp = table.to_pandas()
    dfs.append(df_temp)
```

```python
df_merged = pd.concat(dfs, ignore_index=True)
print(f"Merged dataset shape: {df_merged.shape}")
```

```python
output_path = "data/cleaned/yellow_taxi_merged_2025.parquet"
df_merged.to_parquet(output_path, index=False)
print(f"Merged dataset saved to {output_path}")
```

```python
memory_MB = df_merged.memory_usage(deep=True).sum() / (1024**2)
print(f"Approximate memory usage: {memory_MB:.2f} MB")
```
